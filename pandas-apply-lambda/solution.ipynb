```python
import pandas as pd
```

```python
df = pd.read_csv('data/input/IMDB-Movie-Data.csv', sep =',')
df.head(10)
```

### CHALLENGE 1

```python
serie = df['Votes']
serie
```

```python
def bin_column(i):
    if i < 1000:
        return 1
    elif i < 10000:
        return 2
    elif i < 100000:
        return 3
    elif i < 1000000:
        return 4
    else:
        return 5

df['bin'] = df['Votes'].apply(bin_column)  
df.head()
```

### CHALLENGE 2

```python
def revenue(millions, minutes):
    return millions/minutes
df['revenue'] = df.apply(lambda x: revenue( x['Revenue (Millions)'], x['Runtime (Minutes)']), axis=1 )
df.head(5)
```

### CHALLENGE 3

```python
genre = ['Action,Adventure,Sci-Fi', 'Animation,Comedy,Family', 'Animation, Thriller, Adventure']
rating = [8.1, 7.0, 7.3]
new = []
for i, x in zip(genre, rating):
    if 'Thriller' in i:
        new.append(x + 1)
    elif 'Comedy' in i:
        new.append(x - 1)
    else:
        pass
print(new)
```

```python
def new_rating (genre, rating):
    new_rating = []
    for i, n in zip(genre, rating):
        if 'Thriller' in i:
            new_rating.append(n + 1)
        elif 'Comedy' in i:
            new_rating.append(n - 1)
        else:
            new_rating.append(n)
    return new_rating
df['new rating'] = df.apply(lambda x: new_rating( x['Genre'], x['Rating']), axis=1 )

df.head(5)
    
```
